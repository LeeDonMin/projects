# 공통 실습 코드

```python
#Step 3. 훈련용 세트와 테스트용 세트로 나눕니다.
# '무게_g','길이_cm','색상','당도'에 따른 과일종류 분류
from sklearn.model_selection import train_test_split
data = src_data[['무게_g','길이_cm','색상','당도']]
target = src_data['종류']

# train, test 데이터 분리
훈련용_data, 테스트용_data, 훈련용_target, 테스트용_target=train_test_split(data,target, test_size=0.2)
```

```python
# 각각의 데이터 확인
print(훈련용_data.shape , 테스트용_data.shape)
print(훈련용_data)
print(훈련용_target)
```

```python
# 랜덤 포레스트 모델 생성
from sklearn.ensemble import RandomForestClassifier
# 학습
rt = RandomForestClassifier()
rt.fit(훈련용_data, 훈련용_target)
# 예측
rt.predict(테스트용_data)
# score
```
