```python
import folium
```

```python
center = [37.541, 126.986] # 지도를 표시할 중심 좌표
m = folium.Map(width=600, height=400, location=center, zoom_start=10)

#
for i in df.index[:1]: #한줄만 찍기 
    folium.Circle(
        location = df.loc[i, ['Latitude', 'Longitude']],
        tooltip = df.loc[i, 'Station'],
        radius = 2000   # 원의 크기
    ).add_to(m)

    print(df.loc[i]['Station'])
m
```
