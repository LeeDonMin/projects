```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import numpy as np
import pandas as pd
import time
headers ="Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/127.0.0.0 Safari/537.36 Edg/127.0.0.0"
# 브라우저 꺼짐 방지 옵션
chrome_options = Options()
chrome_options.add_experimental_option("detach", True)
chrome_options.add_argument(f'--user-agent={headers}')
# chrome_options.add_argument("headless")
driver = webdriver.Chrome(options=chrome_options)
driver.implicitly_wait(15)
url='https://www.capology.com/club/west-ham/payrolls/'
driver.get(url)
tester=driver.find_elements(By.XPATH,'/html/body/nav/ul/li[3]/div[2]/div/ul/li[1]/div[2]/div/ul/li[1]/div[2]/div/div[5]/select')

for team in tester:
    if 'text' in dir(team):
        print(team.text)
        


teamnamelist=team.text.lower().replace(' ','-').split('\n')[1:]
print(teamnamelist)
driver.quit()
epllist=[]
for teamname in teamnamelist:
    headers ="Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/127.0.0.0 Safari/537.36 Edg/127.0.0.0"
    chrome_options = Options()
    chrome_options.add_argument(f'--user-agent={headers}')
    chrome_options.add_argument("headless")
    url=f'https://www.capology.com/club/{teamname}/payrolls/'

    teamdriver = webdriver.Chrome(options=chrome_options)
    teamdriver.implicitly_wait(15)
    teamdriver.get(url)
    tester2=teamdriver.find_elements(By.XPATH,'/html/body/main/div[3]/div/div[6]/div[2]/div[3]/div[2]/table')

    
    for get2 in tester2:
        paylist = get2.text.split('\n')
        columns=paylist[:3]
        indexs=paylist[3:5]
        column=[paylist[3],paylist[4]," ".join(paylist[5:7])," ".join(paylist[7:9])," ".join(paylist[9:11])," ".join(paylist[11:13])," ".join(paylist[13:15]),
                " ".join(paylist[15:17])," ".join(paylist[17:19])]
        values=paylist[19:]
        print(values)
        valuelist=[]

        i=0
        for value in values[:-9]:
            value=value.split(' £ ')
            set=[value[0][:-9].strip(),value[0][-9:].strip()]
            value[0]=set[0]
            value.insert(1,set[1])
            for x in range(len(value[2:])):
                value[2+x]=int("".join(value[2+x].split(',')))
            valuelist.append(value)
        teamchart=pd.DataFrame(data=valuelist,columns=column,index=[teamname]*len(valuelist))
        epllist.append(teamchart)
        teamdriver.quit()
        time.sleep(5)
EPL=pd.concat(epllist)
```
